```python
import folium
```

```python
m = folium.Map([47.53460, 7.641994], zoom_start=18)
m
```

```python
m = folium.Map([47.53460, 7.641994], zoom_start=18, tiles="Stamen Toner")
m
```

```python
m = folium.Map([47.53460, 7.641994], zoom_start=18, tiles="OpenStreetMap")

fhnw = [47.534832443156034, 7.641978798605998]
bahnhof = [47.53369088502312, 7.647735469448107]
interdiscount = [47.52740982278591, 7.6480757998815925]

folium.Marker(fhnw, popup="Fachhochschule Nordwestschweiz FHNW<br><b>Hier ist die FHNW</b>").add_to(m)
folium.Marker(bahnhof, popup="Muttenz Bahnhof").add_to(m)
folium.Marker(interdiscount, popup="Interdiscount").add_to(m)

m.save("muttenz.html")

m
```

```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("geodata/packages/natural_earth_vector.gpkg",
                    layer="ne_10m_admin_0_countries", encoding="utf-8")
```

```python
gdf
```

```python
result = gdf.query("NAME == 'Switzerland' ")
```

```python
result.plot()
```

```python
m = folium.Map([47.53460, 7.641994], zoom_start=10)

folium.GeoJson(result.to_json()).add_to(m)

m
```

```python
m = folium.Map([47.53460, 7.641994], zoom_start=10)


folium.Marker(fhnw, popup="Fachhochschule Nordwestschweiz FHNW<br><b>Hier ist die FHNW</b>").add_to(m)
folium.Marker(bahnhof, popup="Muttenz Bahnhof").add_to(m)
folium.Marker(interdiscount, popup="Interdiscount").add_to(m)

def func(feature):
    return {'fillColor': "#ffff00",
            'color': '#000000'}

folium.GeoJson(result.to_json(), style_function=func).add_to(m)

m.save("Switzerland.html")

m
```
